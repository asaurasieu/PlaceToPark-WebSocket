```python
from ultralytics import YOLO
import os
import yaml
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import glob
```

```python
print("Loading dataset configuration...")
with open("Car-Model-v4.yolov8/data.yaml", "r") as f:
    data_config = yaml.safe_load(f)
    print(f"Dataset classes: {data_config['names']}")
    print(f"Number of classes: {len(data_config['names'])}")
    class_names = data_config['names']
```

```python
print("\nStarting model training...")
model = YOLO("yolov8m.pt") 
dataset = "Car-Model-v4.yolov8/data.yaml"
results = model.train(
    data=dataset,
    epochs=50,
    imgsz=640,
    batch=16,
    name="parking_detection_model"
)
```
